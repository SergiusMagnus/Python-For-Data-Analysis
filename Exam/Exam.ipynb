```python
import zipfile

import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

with zipfile.ZipFile("archive.zip") as z:
    with z.open("GlobalLandTemperaturesByMajorCity.csv") as f:
        df = pd.read_csv(f, parse_dates=["dt"])

df.head()
```

```python
import numpy as np

#df['NumLatitudeG'] = pd.to_numeric(df['Latitude'].str[:-3]) * np.where(df['Latitude'].str[-1] == 'N', 1, -1)
#df['NumLongitudeG'] = pd.to_numeric(df['Longitude'].str[:-3]) * np.where(df['Longitude'].str[-1] == 'E', 1, -1)

#df['NumLatitudeH'] = pd.to_numeric(df['Latitude'].str[-3:-1]) * np.where(df['Latitude'].str[-1] == 'N', 1, -1)
#df['NumLongitudeH'] = pd.to_numeric(df['Longitude'].str[-3:-1]) * np.where(df['Longitude'].str[-1] == 'E', 1, -1)

#df['Latitude'] = df['NumLatitudeG'] + df['NumLatitudeH'] / 60
#df['Longitude'] = df['NumLongitudeG'] + df['NumLongitudeH'] / 60

df['Latitude'] = pd.to_numeric(df['Latitude'].str[:-1]) * np.where(df['Latitude'].str[-1] == 'N', 1, -1)
df['Longitude'] = pd.to_numeric(df['Longitude'].str[:-1]) * np.where(df['Longitude'].str[-1] == 'E', 1, -1)

df.head()
```

```python
df = df.loc[(df['dt'] >= '1970-01-01') & (df['dt'] <= '1980-01-01')]
df = df[['dt', 'AverageTemperature', 'City', 'Latitude', 'Longitude']]
df.head()
```

```python
df = df.loc[df['dt'].dt.month == 1].groupby(['City']).agg(AverageTemperature=('AverageTemperature', 'mean'), Latitude=('Latitude', 'mean'), Longitude=('Longitude', 'mean'))
df.head()
```

```python
df['City'] = df.index
df = df.merge(right=df, how='cross')
df
```

```python
df.loc[np.modf(df['Latitude_x'])[0] < 10]
```

```python
df['Latitude_x'] = df['Latitude_x'] * (np.pi / 180)
df['Longitude_x'] = df['Longitude_x'] * (np.pi / 180)
df['Latitude_y'] = df['Latitude_y'] * (np.pi / 180)
df['Longitude_y'] = df['Longitude_y'] * (np.pi / 180)
df
```

```python
# df['Distance'] = np.sqrt((df['Latitude_x'] - df['Latitude_y']) ** 2 + (df['Longitude_x'] - df['Longitude_y']) ** 2)
df['Distance'] = np.arcsin(np.sin((df['Latitude_y'] - df['Latitude_x']) / 2) ** 2 + np.cos(df['Latitude_x']) * np.cos(df['Latitude_y']) * np.sin(df['Longitude_y'] - df['Longitude_x']) ** 2) * 6371
df
```

```python
df['Temp'] = np.abs(df['AverageTemperature_x'] - df['AverageTemperature_y'])
df
```
